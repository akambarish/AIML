```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.fit_transform(X_test)
```

```python
X_test
```

```python
def logi_reg (X_train , Y_train):
    from sklearn.linear_model import LogisticRegression
    Logi=LogisticRegression( random_state=0)
    Logi.fit(X_train, Y_train)
    print("Logistic Rgression : Training Acurancy", Logi.score(X_train, Y_train))
    return Logi    
```

```python
logrex = logi_reg(X_train, Y_train)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(Y_test, logrex.predict(X_test))
print(cm)
```

```python
TP=cm[0][0]
TN=cm[1][1]
FN=cm[1][0]
FP=cm[0][1]
print("Testing accuracy of logistic regression model=", (TP+TN)/(TP+TN+FN+FP))
```

```python
def Deci_Tree (X_train , Y_train):
    from sklearn.tree import DecisionTreeClassifier
    DTData = DecisionTreeClassifier(max_depth = 2, random_state=50)
    DTData.fit(X_train, Y_train)
    print("Decision Tree Classifier : ", DTData.score(X_train, Y_train))
    return DTData    
```

```python
Decision_Tree_Data = Deci_Tree(X_train, Y_train)
```

```python
from sklearn.metrics import confusion_matrix
dt_cm = confusion_matrix(Y_test, Decision_Tree_Data.predict(X_test))
print(dt_cm)
```

```python
DTTP=dt_cm[0][0]
DTTN=dt_cm[1][1]
DTFN=dt_cm[1][0]
DTFP=dt_cm[0][1]
print("Testing accuracy of logistic regression model=", (DTTP+DTTN)/(DTTP+DTTN+DTFN+DTFP))
```
